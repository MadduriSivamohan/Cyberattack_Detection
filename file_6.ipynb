```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('finalfile.csv')
```

```python
from sklearn.model_selection import train_test_split
```

```python
max_distribution = df[' Label'].value_counts().max()

# Calculate the number of samples to keep for each label
samples_per_label = {}
for label, count in df[' Label'].value_counts().items():
    samples_per_label[label] = min(count, max_distribution)

# Create an empty DataFrame to store the sampled data
sampled_df = pd.DataFrame(columns=df.columns)

# Stratified sampling for each label
for label in samples_per_label.keys():
    label_data = df[df[' Label'] == label]
    sampled_data = label_data.sample(n=samples_per_label[label], random_state=42)
    sampled_df = pd.concat([sampled_df, sampled_data], axis=0)  # Use concat to combine DataFrames

```
